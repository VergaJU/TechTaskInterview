```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
logs = pd.read_csv('../logs/classifier_training_logs.csv', index_col=0)
```

```python
sns.lineplot(logs)
```

```python
with open('../logs/classifier_metrics.pkl','rb')as f:
    metrics=pickle.load(f)

labels=['Basal-G3', 'Her2-G2', 'Her2-G3', 'LumA-G2', 'LumB-G3', 'Normal-G2']
```

```python
print(metrics['classification_report'])
```

```python
cm=pd.DataFrame(metrics['confusion_matrix'], index=labels, columns=labels)
```

```python
plt.figure()
sns.heatmap(cm, annot=True)
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```
